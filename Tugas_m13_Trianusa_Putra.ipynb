# Tugas 1

```
def cos(frek,tipe,ht,hr,d):
  #frek = frekuensi (MHz)
  #tipe = urban, suburban, open
  #ht = tinggi tower (meter)
  #hr = tinggi user (meter)
  #d = jarak (meter)
    if frek in range(150,1501):
        Ca=69.55
        Cb=26.16
    elif frek in range(1501,2001):
        Ca=46.3
        Cb=33.9
    
    if tipe=='urban' :
        Cm=0
    elif tipe=='suburban' :
        Cm= -2*((np.log10((frek/28)**2))-5.4)
    elif tipe=='open' :
        Cm= -4.78*(np.log10(frek))+18.33*(np.log10(frek))-40.94
    else :
        Cm='tipe salah'
    
    ahr = (1.1*np.log10(frek)-0.7)*hr-(1.56*np.log10(frek)-1.8)
    Lp = Ca+(Cb*np.log10(frek))-(13.83*np.log(ht))-ahr+((44.9-6.55*np.log10(ht))*np.log10(d))-Cm
    
    return Lp
```

```
f=1800 #MHz
loc1='urban'
loc2='suburban'
loc3='open'
tower=80 #m
user=1.5 #m
dmin=200 #m
dmax=700 #m
N=200 #titik
d=np.random.randint(dmin,dmax,N)
```

```
LpUrban=cos(f,loc1,tower,user,d)
LpSuburban=cos(f,loc2,tower,user,d)
LpOpen=cos(f,loc3,tower,user,d)

print(LpUrban)
print(LpSuburban)
print(LpOpen)
```

```
plt.figure()
plt.plot(d,LpUrban, 'b', label='Urban')
plt.plot(d,LpSuburban, 'y', label='Suburban')
plt.plot(d,LpOpen, 'g', label='Open')
plt.legend()
plt.title('FREKUENSI 1800 MHZ')
plt.xlabel('Jarak (m)')
plt.ylabel('Path Loss (dB)')
plt.show()
```

```
d_sort=d[np.argsort(d)]
Urban_sort=LpUrban[np.argsort(d)]
Suburban_sort=LpSuburban[np.argsort(d)]
Open_sort=LpOpen[np.argsort(d)]
```

```
plt.figure()
plt.plot(d_sort,Urban_sort, 'b', label='Urban')
plt.plot(d_sort,Suburban_sort, 'y', label='Suburban')
plt.plot(d_sort,Open_sort, 'g', label='Open')
plt.legend()
plt.title('FREKUENSI 1800 MHZ')
plt.xlabel('Jarak (m)')
plt.ylabel('Path Loss (dB)')
plt.show()
```
